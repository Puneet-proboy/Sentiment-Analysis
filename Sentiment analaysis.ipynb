```python
# Making dict of abbrivation symbol and names of companies
details_of_comapnies = comapanies_in_america.set_index('Stock Abbreviation')['Company Name'].to_dict()
details_of_comapnies
```

```python
# # Making new dataset includes ['Comapny Name','Short Form', 'Titles']
sentiment_data = pd.DataFrame(columns =  ['Comapny Name','Short Form', 'Titles'])


# improting each news form the stock abbrivations 


for abbrivation_companies,name_companies in details_of_comapnies.items():
    
    articles = news.get_yf_rss(abbrivation_companies)  #taking news frm the yahoo finance spesefic to that companies
                                         
    
    # loop for taking the title out from the news 
    for main_heading in articles:
        #print the headind along the name of the coampanies
        
        
        # now we are making new dataset for sentimental analyies on the news of the each comapnies 
        sentiment_data = sentiment_data.append(
            {
                'Comapny Name' : name_companies,
                'Short Form' : abbrivation_companies,
                'Titles'  : main_heading['title']
            }, ignore_index = True) # ignoring the error
    
```

# Removing stopwords or cleaning 

```python
import re # for removing special char

def data_cleaning(letter: str): # defing the function for cleainng the text 
    letter_token = word_tokenize(letter.lower())
    
    # variable to remove the stopwords form english language
    stopwords_remove = stopwords.words('english')
    
    # removing titles stopwords
    removing_titles_stopwords = [letter for letter in letter_token if letter not in stopwords_remove]
   
    # for lemmitizing the text 
    lemmatize_words = WordNetLemmatizer()
    lemmatizie_title = [lemmatize_words.lemmatize(letter) for letter in removing_titles_stopwords]
    
    # removing special char while returning
    return re.sub(r'[^a-zA-Z0-9\s]','',' '.join(lemmatizie_title))
    
    
```

```python
# Unit testing
data_cleaning(sentiment_data['Titles'][0]) 
```

# Performing sentimnetal analysis

```python
def news_sentiment(text : str):#function to check thet the setiment id postive or nagtive
    analyzer = SentimentIntensityAnalyzer()
    
    checking_sentiment  = analyzer.polarity_scores(text)
    
    if checking_sentiment['compound'] >= 0.05: # for postive responce
        return 1
    elif checking_sentiment['compound'] <= -0.05: # for negative response
        return -1
        
    else:
        return 0 # for netural response
```

```python
# news_sentiment(data_cleaning(sentiment_data['Titles']))
```

```python
list_sentiments = []
for title in sentiment_data['Titles']:      #making list of the sentimental analysis 
    list_sentiments.append(news_sentiment(data_cleaning(title)))
    
```

```python
# cheching all the senitemt outupts
print(list_sentiments)
```

```python
# adding the sentiment list as a column to our dataset
sentiment_data['Sentiments'] = list_sentiments 
```
